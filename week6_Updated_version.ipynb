```python
contacts = {
    "Prakriti": {"phone": "0412 345 678", "email": "prakriti@example.com"},
    "Nimisha": {"phone": "0423 567 890", "email": "Nimisha@example.com"},
    "Jenisha": {"phone": "0434 678 901", "email": "jenisha@example.com"}
}

contacts["Ram"] = {"phone": "0455 789 012", "email": "ram@example.com"}

print("Contact Details:\n")

for name, info in contacts.items():
    print(name)
    print(" Phone:", info["phone"])
    print(" Email:", info["email"])
    print()

    #OPTIONAL lookup
search = input("Enter name: ")

if search in contacts:
    print("\nContact Found:")
    print("Phone:", contacts[search]["phone"])
    print("Email:", contacts[search]["email"])
else:
    print("Contact not found")
```

My Prompt to AI:

I am a beginner in Python. I have created a simple dictionary for contacts with names and phone numbers. How can I modify it so that each contact can also store an email? Please keep it simple and easy to understand.

AI Reply:

You can use a nested dictionary. Instead of storing just a phone number as the value, you can store another dictionary that contains both phone and email.

For example:

contacts = {
    "Alice": {"phone": "0400 123 456", "email": "alice@example.com"},
    "Bob": {"phone": "0411 000 999", "email": "bob@example.com"}
}

This way, each contact can store multiple pieces of information. You can access them like:
contacts["Alice"]["phone"]
contacts["Alice"]["email"]

To print all contacts, you can loop using .items():
for name, info in contacts.items():
    print(name, info["phone"], info["email"])
