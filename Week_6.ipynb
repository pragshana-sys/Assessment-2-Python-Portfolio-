```python
# Python script to form a dictionary named " grade" with keys of programming ,Math ,Account, Science and values of 8,5,3,7 then print the pairs

grade = {
    "Programming": 8,
    "Math": 5,
    "Account": 3,
    "Science": 7
}

for subject, value in grade.items():
    print(subject, ":", value)
```

```python
#Creating a small dictionary for grades in 3 subjets , then : 1.Add a new subject 2. Update one grade 3. Print all subjects and grades using .items()

grades = {
    "Math": 5,
    "Accounts": 0,
    "Pragramming": 8
}
print("Before change :\n" ,grades)

# 1.Here, Adding a new subject :
grades["History"] = 8

# 2. Here,Updating one grade :
grades["Accounts"] = 6

# 3.Now, Printing all subjects and grades using .items() :
print("\nAfter change ")
for subject, grade in grades.items():
    print(subject, ":", grade)
```

```python
#In class-Design challenge: In pairs, design a nested dictionary structure for (there were three options among which I chose RESTAURANT MENU)

menu = {
    "Starters": {
        "Spring Rolls": 6,
        "Garlic Bread": 5
    },
    "Main Course": {
        "Chicken Curry": 12,
        "Veg Pasta": 10
    },
    "Drinks": {
        "Coke": 3,
        "Juice": 4
    }
}

for category, items in menu.items():
    print("\n" + category)
    for item, price in items.items():
        print(item, ":", price)
```
